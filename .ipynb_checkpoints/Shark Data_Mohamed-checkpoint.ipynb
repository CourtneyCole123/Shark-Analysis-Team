```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
import json
```

```python
base_url = 'https://public.opendatasoft.com/api/explore/v2.1/catalog/datasets/global-shark-attack/records?limit=100'
```

```python
requests.get(base_url)
```

```python
#request = (requests.get(base_url).json())
```

```python
total_count = requests.get(base_url).json()['total_count']
```

```python
#hold the data
all_of_the_data = []
```

```python
for currentOffset in np.arange(0,total_count,100):
  # print(currentOffset)
  cur_data = requests.get(f'{base_url}&offset={currentOffset}').json()['results']
  all_of_the_data.extend(cur_data) #extend is like
  time.sleep(0.5)
```

```python
shark_clean_data_df.head(3)
```

```python
# Subset data based on country, location and area to focus on analysis based on countries
shark_clean_data_Geo_df = shark_clean_data_df[["country", "location","area"]]
shark_clean_data_Geo_df.head()
```

```python
# list the top countries and number of shark attack incidences

shark_clean_data_Geo_df[['country','area']].value_counts().head()
```

```python
# group data by country and use for plotting
Attack_per_country = shark_clean_data_Geo_df.groupby('country')

# Count how many times each road type appears in our group
Attack_per_country = Attack_per_country['country'].count()

Attack_per_country.count()
```

```python
# use bar graph to plot number of shark attack per country
count_chart = Attack_per_country.plot(kind='bar', figsize=(6,8), title="Attacks per country")

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("country")
count_chart.set_ylabel("Number of Attacks")
# tick_locations = [value for value in country]
# plt.xticks(tick_locations, country)
# plt.show()
plt.tight_layout()
plt.savefig("Bargraph_SharkAttack.png")
plt.show()
```

```python
# Subset the top countries in shark attack incidences
Top_countries_df =shark_clean_data_Geo_df['country'].value_counts().head()
Top_countries_df
```

```python
# Create a pie chart based upon the proporportion of shark attack in top shark incident countries
explode = (0.1,0,0,0,0)
attack_pie = Top_countries_df.plot(kind="pie", explode=explode,autopct="%1.1f%%", shadow=True)
attack_pie.set_ylabel("country attacks")
plt.ylabel("")
plt.axis("equal")
# Save an image of our chart and print the final product to the screen
plt.savefig("PyPies_SharkAttack.png")
plt.show()
```
